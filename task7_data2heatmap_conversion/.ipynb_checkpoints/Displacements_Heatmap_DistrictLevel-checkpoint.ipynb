### Somalian geodata

```python
dfg = gpd.read_file("geodata_somalia/som_admbnda_adm2_undp.shp")
dfg = dfg.sort_values(by='admin2Name')

dfg = dfg[['admin2Name','admin1Name','admin0Name', 'Shape_Area','geometry']]

dfg.loc[dfg[dfg['admin2Name'] == 'Baydhaba'].index,'admin2Name'] = 'Baidoa'
dfg.loc[dfg[dfg['admin2Name'] == 'Kismaayo'].index,'admin2Name'] = 'Kismayo'

dfg['regdis'] = dfg['admin1Name'] + ' - ' + dfg['admin2Name'] 
```

```python
# Read displacements
dpath = {}
dpath['home'] = './data/displacements'
dpath['subfolder'] = 'input'
dpath['filename'] = 'UNHCR-PRMN-Displacement-Dataset.xlsx'
dfd = pd.read_excel(os.path.join(dpath['home'],dpath['subfolder'],dpath['filename']))

rd_names = pd.DataFrame(columns = ['Region','District'])

x1 = dfd[['Current (Arrival) Region','Current (Arrival) District']]
x1.columns = ['Region','District']
x2 = dfd[['Previous (Departure) Region','Previous (Departure) District']]
x2.columns = ['Region','District']

rd_names = rd_names.append(x1)
rd_names = rd_names.append(x2)

rd_names = rd_names.drop_duplicates(subset=['Region','District']).sort_values(by=['Region','District'])
rd_names['regdis'] = rd_names['Region'] + ' - '+ rd_names['District']
# Find mismatches between the district names

dfg['regdis'] = dfg['admin1Name'] + ' - ' + dfg['admin2Name'] 
dfg['regdis'].append(rd_names['regdis']).drop_duplicates(keep = False)

# There are no more mismatches between the two datasets
# These remaining values are simply districts in which no displacements are recorded
```

```python
dfg = dfg[['regdis','geometry','Shape_Area']]
dfg.head()
```

```python
dfg.sort_values(by='regdis').plot()
```

# Displacement data

```python
dfd.head()
```

```python
from datetime import datetime
dfd['ts'] = [datetime.strptime(str(i) + '0', "%Y%W%w") for i in dfd['Year Week']]
```

```python
# We want to predict drought related displacements
dfd = dfd[dfd['Reason'] == 'Drought related']
```

# Arrivals

```python
dfs = {}
df = dfd[['ts','Current (Arrival) Region','Current (Arrival) District','Number of Individuals']]
df.columns = ['ts','region','district','individuals']
df['regdis'] = df['region'] + ' - '+ df['district']
df = df.groupby(['regdis','ts'],as_index=False).sum()
df = df.pivot(index='ts',columns='regdis')
df = df.fillna(0)
df.columns = [i[1] for i in list(df.columns)]

dfs['arrivals'] = df
```

```python
df.head()
```

```python
dpath = {}
dpath['home'] = './data/displacements'
dpath['subfolder'] = 'output'
dpath['filename'] = 'Somalia-Displacements-Arrivals-2016-2019-DistrictLevel.csv'
df.to_csv(os.path.join(dpath['home'],dpath['subfolder'],dpath['filename']))
```

# Departures

```python
df = dfd[['ts','Previous (Departure) Region','Previous (Departure) District','Number of Individuals']]
df.columns = ['ts','region','district','individuals']
df['regdis'] = df['region'] + ' - '+ df['district']
df = df.groupby(['regdis','ts'],as_index=False).sum()
df = df.pivot(index='ts',columns='regdis')
df = df.fillna(0)
df.columns = [i[1] for i in list(df.columns)]

dfs['departures'] = df
```

```python
df.head()
```

```python
dpath = {}
dpath['home'] = './data/displacements'
dpath['subfolder'] = 'output'
dpath['filename'] = 'Somalia-Displacements-Departures-2016-2019-DistrictLevel.csv'
df.to_csv(os.path.join(dpath['home'],dpath['subfolder'],dpath['filename']))
```

# Visualization

```python
arive_or_depart = 'arrivals'
# arive_or_depart = 'departures'

for ind in range(0,10):
    tsi = dfs[arive_or_depart].index[ind]

    x = pd.DataFrame(dfs[arive_or_depart].loc[tsi,:].reset_index())
    x.columns=['regdis','individuals']

    # Plot settings
    fig, ax = plt.subplots(figsize=(8,8), subplot_kw={'aspect':'equal'})
    ax.set_xlim(40, 52.5)
    ax.set_ylim(-2.5,12.5)
    #ax.set_axis_off()
    imtitle = arive_or_depart + ' ' + str(tsi)
    ax.set_title(imtitle, size = 20)

    # Plot satellite data on top of natural earth
    dfg.merge(x).plot(ax=ax, column='individuals', cmap='coolwarm', legend=True);

    dpath = {}
    dpath['home'] = './data/displacements'
    dpath['subfolder'] = 'images'
    dpath['filename'] = imtitle

    fig.savefig(os.path.join(dpath['home'],dpath['subfolder'],dpath['filename']))
```
